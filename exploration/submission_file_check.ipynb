```python
# submission 1
file_path = '../submissions/submission01/LP1d_PredictionChallenge.csv'


labels = pd.read_csv(file_path)
labels['duration'] = labels.offset - labels.onset
# histogram
labels.groupby('event_label').hist(column='duration', bins=100)
# calculate how often an event appears on average
plt.clf()
current_file = '00001_mix.wav'
counts = [0]
for file in labels['filename']:
    if file != current_file:
        counts.append(0)
        current_file = file
    counts[-1] += 1
plt.hist(counts, bins=100)
plt.show()
plt.clf()
labels_agg = labels.groupby('event_label')['duration'].sum()
labels_agg.plot.bar()
plt.show()
```
