```python
import pandas as pd
import numpy as np
```

```python
trainDf = pd.read_csv('../data/train.csv')
trainDf['duration'] = trainDf.offset - trainDf.onset
valDf = pd.read_csv('../data/validation.csv')
valDf['duration'] = valDf.offset - valDf.onset
testDf = pd.read_csv('../data/test.csv')
testDf['duration'] = testDf.offset - testDf.onset
challengeDf = pd.read_csv('../submissions/submission01/LP1d_PredictionChallenge.csv')
challengeDf['duration'] = challengeDf.offset - challengeDf.onset
testPred = pd.read_csv('../trainings/LP1d_Prediction.csv')
testPred['duration'] = testPred.offset - testPred.onset
```

```python
train_agg = trainDf.groupby('event_label')['duration'].sum()
train_agg.plot.bar()
```

```python
val_agg = valDf.groupby('event_label')['duration'].sum()
val_agg.plot.bar()
```

```python
test_agg = testDf.groupby('event_label')['duration'].sum()
test_agg.plot.bar()
```

```python
chall_agg = challengeDf.groupby('event_label')['duration'].sum()
chall_agg.plot.bar()
```

```python
testPred_agg = testPred.groupby('event_label')['duration'].sum()
testPred_agg.plot.bar()
```
